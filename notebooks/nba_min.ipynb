# NBA Cleaning â€” Minimal Version

```

import csv, numpy as np, pandas as pd
from pathlib import Path

p = Path("nba.csv")  # put the csv next to the notebook
print(p.resolve())
```

```

# 1) lines -> array
file_array = p.read_text(encoding="utf-8", errors="ignore").splitlines()
len(file_array), file_array[:2]
```

```

# 2) array -> rows -> DataFrame
rows = list(csv.reader(file_array))
raw = pd.DataFrame(rows[1:], columns=rows[0])
raw.head()
```

```

# 3) UDF
def clean(x):
    if x is None: return np.nan
    s = str(x).strip()
    if s == "" or s.lower() in {"na","n/a","nan","none","null","--"}: return np.nan
    t = s.replace("$","").replace(",","").replace("%","")
    try:
        u = t.replace("-","",1).replace(".","",1)
        if u.isdigit():
            v = float(t); return int(v) if v.is_integer() else v
    except: pass
    return s
```

```

# 4) applymap + save
cleaned = raw.applymap(clean)
for c in cleaned.columns: cleaned[c] = pd.to_numeric(cleaned[c], errors="ignore")
print(raw.dtypes, "\n---\n", cleaned.dtypes)
cleaned.head()
```

```

cleaned.to_csv("nba_clean.csv", index=False); print("Saved nba_clean.csv")
```
